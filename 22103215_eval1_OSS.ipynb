```python
books={}

def add_book(ISBN, title, author, genre, availability_status):
    if ISBN in books:
        print("A book with similar ISBN already exists.")
        return
    books[ISBN]={
#         "ISBN": ISBN,
        "title": title,
        "author": author,
        "genre": genre,
        "availability_status": availability_status
    }

def update_book_details(ISBN, title=None, author=None, genre=None, availability_status=None):
    if ISBN not in books:
        print("Invalid ISBN")
        return None
    if title:
        books[ISBN]["title"]=title
    if author:
        books[ISBN]["author"]=author
    if genre:
        books[ISBN]["genre"]=genre
    if availability_status:
        books[ISBN]["availability_status"]=availability_status
    return books[ISBN]
        
        
def search_by_isbn(ISBN):
    if ISBN in books:
        return books[ISBN]
    else:
        print("No book found with given ISBN")
        return None
    
def  generate_genre_report():
    genreBooks={}
    for key,value in books.items():
        if value["genre"] in genreBooks:
            genreBooks[value["genre"]].append(value)
        else:
            genreBooks[value["genre"]]=[value]
    return genreBooks

add_book("1001", "My Life", "Vanshul Agarwal", "Autobiography", True)
add_book("1002", "The Flamingson", "Richard Flamingson", "Autobiography", True)
add_book("1003", "Think and Grow Rich", "Satya Nadela", "Self-improvement", False)
add_book("1004", "Rich Dad Poor Dad", "Harvy Specter", "Self-improvement", True)
add_book("1005", "It Ends With Us", "Haley Coover", "Fictional", True)
add_book("1006", "It Starts With Us", "Haley Coover", "Fictional", False)
print("ORIGINAL DICTIONARY: ")
print(books,"\n")

isbn=input("Enter ISBN: ")
title=input("Enter title: ")
author=input("Enter author: ")
genre=input("Enter genre: ")
avail=bool(input("Enter availability status 0 for False 1 for True: "))
add_book(isbn, title, author, genre, avail)
print("DICTIONARY: ")
print(books,"\n")

print("UPDATE A BOOK")
isbn=input("Enter ISBN: ")
title=input("Enter title: ")
author=input("Enter author: ")
genre=input("Enter genre: ")
avail=bool(input("Enter availability status 0 for False 1 for True: "))
update_book_details(isbn, title, author, genre, avail)
print(books, "\n")

print("SEARCH:")
x=input("Enter ISBN: ")
print(search_by_isbn(x), "\n")

print("GENRE REPORT: ")
genreReport = generate_genre_report()

for genre, items in genreReport.items():
    print(genre)
    for i in items:
        print(i)
    print()
```
